```python
import pandas as pd
import numpy as np
import tables
from tqdm import *
```

```python
test_pred10 = np.load('test_pred_prob_10neigh.npy')
full_test_ind10 = np.load('test_inds_prob_10neigh.npy')
full_test10 = np.load('full_test_prob_10neigh.npy')

test_pred6 = np.load('test_pred_prob_6neigh.npy')
full_test_ind6 = np.load('test_inds_prob_6neigh.npy')
full_test6 = np.load('full_test_prob_6neigh.npy')

test_pred3 = np.load('test_pred_prob_3neigh.npy')
full_test_ind3 = np.load('test_inds_prob_3neigh.npy')
full_test3 = np.load('full_test_prob_3neigh.npy')
```

```python
preds = [[test_pred10, full_test_ind10, full_test10],
        [test_pred6, full_test_ind6, full_test6],
        [test_pred3, full_test_ind3, full_test3]]
```

```python
all_data_models = []

for test_pred, full_test_ind, full_test in tqdm(preds):
    x = full_test[::, 1].astype(np.int32)[::, np.newaxis]
    y = full_test[::, 2].astype(np.int32)[::, np.newaxis]
    z = full_test[::, 3].astype(np.int32)[::, np.newaxis]
    all_data_models.append(np.hstack((full_test_ind[::, np.newaxis], x, y, z, test_pred)))
```

```python
from collections import defaultdict
def dd():
    return defaultdict(lambda: 1)

full_ans = defaultdict(dd)

for cur_model in all_data:
    for ind, x, y, z, *probs in tqdm(cur_model):
        ind, x, y, z = int(ind), int(x), int(y), int(z)
        full_ans[ind][(x, y, z)] *= np.array(probs)
```

```python
def rlencode(x, dropna=False):
    where = np.flatnonzero
    x = np.asarray(x)
    n = len(x)
    if n == 0:
        return (np.array([], dtype=int), 
                np.array([], dtype=int), 
                np.array([], dtype=x.dtype))

    starts = np.r_[0, where(~np.isclose(x[1:], x[:-1], equal_nan=True)) + 1]
    lengths = np.diff(np.r_[starts, n])
    values = x[starts]
    
    if dropna:
        mask = ~np.isnan(values)
        starts, lengths, values = starts[mask], lengths[mask], values[mask]
    
    return starts, lengths, values
```
